```python
import matplotlib.pyplot as plt 
from dataclasses import dataclass
from typing import List
import numpy as np
%matplotlib inline 
```

```python
@dataclass 
class PruningResult:
    
    levels: List[float]
    general_accuracies: List[float]
    bus_accuracies: List[float]
    car_accuracies: List[float]
    dog_accuracies: List[float]
    motorbike_accuracies: List[float]
    person_accuracies: List[float]
    train_accuracies: List[float]
        
    description: str
        
    def __init__(self,levels: List[float],
                 general_accuracies: List[float],
                 bus_accuracies: List[float],
                 car_accuracies: List[float],
                 dog_accuracies: List[float],
                 motorbike_accuracies: List[float],
                 person_accuracies: List[float],
                 train_accuracies: List[float],
                 description: str):
        self.general_accuracies = np.dot(general_accuracies,100)
        self.levels = levels
        self.description = description
        self.bus_accuracies = np.dot(bus_accuracies,100)
        self.car_accuracies = np.dot(car_accuracies, 100)
        self.dog_accuracies = np.dot(dog_accuracies, 100)
        self.motorbike_accuracies = np.dot(motorbike_accuracies, 100)
        self.person_accuracies = np.dot(person_accuracies, 100)
        self.train_accuracies = np.dot(train_accuracies, 100)
        
    def plot(self):
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.general_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title(self.description)
        plt.savefig('resources/l1-gen.png')
        
        
        #Bus
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.bus_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title("Bus")
        plt.savefig('resources/l1-bus.png')
        
        
        #Car
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.car_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title("Car")
        plt.savefig('resources/l1-car.png')
        
        
        
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.dog_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title("Dog")
        plt.savefig('resources/l1-dog.png')
        
        
        #Motorbike
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.motorbike_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title("Motorbike")
        plt.savefig('resources/l1-motorbike.png')
        
        
        #Person
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.person_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title("Person")
        plt.savefig('resources/l1-person.png')
        
        
        #Train
        fig = plt.figure()
        ax = plt.axes()
        ax.plot(
            self.levels,self.train_accuracies, linestyle='-', marker='o', color='r'
        )
        ax.set_xticks(self.levels)
        plt.xlabel("Pruning Percentage")
        plt.ylabel("Accuracy")
        plt.title("Train")
        plt.savefig('resources/l1-train.png')
        
```
